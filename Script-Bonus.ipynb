```python
#!/usr/bin/python

import os, sys
import json
from pprint import pprint
# Open a file
path = "./jsons_Mensual/"
dirs = os.listdir( path )

config = {}

# This would print all the files and directories
for file in dirs:
    f = open(path + file)
    data = json.load(f)
    config[data["table_name"]] = data
    f.close()
```

```python
pprint(config)
```

```python
with open("./json_Consolidado/consolidado.json", "w") as outfile:
    json.dump(config, outfile)
    outfile.close()
```
